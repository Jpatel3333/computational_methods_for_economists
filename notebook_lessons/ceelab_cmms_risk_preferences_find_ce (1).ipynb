# Algorithm pseudo code:


```
L = lottery
U = utility of money
low_v = lowest value outcome of L
EU = expected utility of L
EV = expected value of L
e = some small epsilon like .001

ub = EV      #45 in example
lb = low_v   # 0 in example

while(True):  start an infinite loop
    guess = (ub + lb)/2.  #this is shown in picture for first iteration
    if U(guess) > (eu - e) and U(guess) < (eu + e):
        return guess   # found CE up to e tolerance
    else:
        if U(guess) > eu:
            ub = guess
        else:
            lb = guess
```
